```python
import numpy as np
from otocs.circuit import (
    sweep_echo_k,
)
from otocs.model import (
    ChaoticModel
)
```

```python
num_qubit = 10
obs = ChaoticModel(num_qubit)
obs.add_params(
    # J_x=-0.2,
    # J_y=-0.0,
    # J_z=-0.0,
    # h_z=0.5,
)
observable=obs.get_observable()

results = sweep_echo_k(
    observable=observable,
    echo_k_range=[1/2],
    time_range=np.arange(0,4.01,0.1),
    targets=(0,9),
    delta=0.5,
)
```

```python
results.plot()
```
