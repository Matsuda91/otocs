```python
import plotly.graph_objects as go
import plotly.io as pio
from collections import defaultdict
from tqdm import tqdm
import numpy as np
import qulacs as qs
from otocs.model import (
    Observable,
    Lattice
)
```

```python
num_qubit = 10
obs = Observable(num_qubit)
obs.add_chaotic_model(
    J_x=-0.2,
	J_y=-0.0,
    J_z=0.0,
    h_z=-0.1
)
observable=obs.get_observable()
observable.is_hermitian()
```

```python
times = np.arange(0.0, 7.01, 0.2)
```

```python
from otocs.src.fujii_arxiv import (
	compute_fig1_theta_histograms,
	compute_fig2_moments,
)
```

```python
compute_fig1_theta_histograms(
		observable=observable,
		times=times,
		i=0,
		j=9,
	)
```

```python
for j in range(num_qubit):
	try:
		compute_fig1_theta_histograms(
				observable=observable,
				times=times,
				i=0,
				j=j,
			)
	except ValueError as e:
		print(f"j={j}: {e}")
		continue
```

```python

compute_fig2_moments(observable=observable, times=times, i=0, js=list(range(10)), k=1/2)
compute_fig2_moments(observable=observable, times=times, i=0, js=list(range(10)), k=1)
compute_fig2_moments(observable=observable, times=times, i=0, js=list(range(10)), k=2)
compute_fig2_moments(observable=observable, times=times, i=0, js=list(range(10)), k=3)
```
